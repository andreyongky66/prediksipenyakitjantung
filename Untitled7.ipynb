```python
# ======================
# IMPROVED CODE VERSION
# ======================

# 1. Enhanced Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats  # For statistical tests

from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier  # Added RandomForest for comparison
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score
from yellowbrick.classifier import PrecisionRecallCurve, ROCAUC, ConfusionMatrix, ClassPredictionError
from yellowbrick.model_selection import LearningCurve, FeatureImportances


from google.colab import files

uploaded = files.upload()


# 2. Data Loading with More Robust Checks
try:
    df = pd.read_csv('heart.csv')
    print("Dataset loaded successfully!")

    # Data quality check
    print("\n=== Data Quality Report ===")
    print(f"Total Rows: {df.shape[0]}")
    print(f"Total Columns: {df.shape[1]}")
    print("\nMissing Values:")
    print(df.isna().sum())
    print("\nDuplicate Rows:", df.duplicated().sum())

except Exception as e:
    print(f"Error loading dataset: {str(e)}")
    exit()

# 3. Enhanced EDA with Statistical Tests
def enhanced_eda(df):
    """Expanded Exploratory Data Analysis with statistical tests"""

    # Set style
    #plt.style.use('seaborn')  # This line is incorrect, causing the error
    sns.set_style('darkgrid')  # Use Seaborn's set_style instead
    colors = ['#D90404', '#F2F2F2']

    # ========== Categorical Features ==========
    cat_features = ['Sex', 'ChestPainType', 'FastingBS', 'RestingECG', 'ExerciseAngina', 'ST_Slope']

    fig = plt.figure(figsize=(20, 15))
    gs = fig.add_gridspec(3, 2)
    plt.suptitle("Enhanced Categorical Feature Analysis", fontweight='heavy', fontsize=24, y=1.02)

    for i, feature in enumerate(cat_features):
        ax = fig.add_subplot(gs[i//2, i%2])

        # Add chi-square test for categorical features
        contingency_table = pd.crosstab(df[feature], df['HeartDisease'])
        chi2, p, _, _ = stats.chi2_contingency(contingency_table)

        sns.countplot(x=feature, hue='HeartDisease', data=df, ax=ax, palette=colors, edgecolor='#000000')
        ax.set_title(f"{feature}\n(Chi2 p-value: {p:.4f})", fontsize=12)
        sns.despine()

    plt.tight_layout()
    plt.savefig('categorical_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

    # ========== Numerical Features ==========
    num_features = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']

    fig = plt.figure(figsize=(20, 15))
    gs = fig.add_gridspec(3, 2)
    plt.suptitle("Enhanced Numerical Feature Analysis", fontweight='heavy', fontsize=24, y=1.02)

    for i, feature in enumerate(num_features):
        ax = fig.add_subplot(gs[i//2, i%2])

        # Add t-test for numerical features
        group0 = df[df['HeartDisease'] == 0][feature]
        group1 = df[df['HeartDisease'] == 1][feature]
        t_stat, p_val = stats.ttest_ind(group0, group1, equal_var=False)

        sns.kdeplot(x=feature, hue='HeartDisease', data=df, fill=True, alpha=0.5,
                   linewidth=0, ax=ax, palette=colors)
        ax.set_title(f"{feature}\n(T-test p-value: {p_val:.4f})", fontsize=12)
        sns.despine()

    plt.tight_layout()
    plt.savefig('numerical_analysis.png', dpi=300, bbox_inches='tight')
    plt.show()

    # Correlation matrix with statistical significance
    plt.figure(figsize=(16, 10))
    corr = df.corr(numeric_only=True)
    mask = np.triu(np.ones_like(corr, dtype=bool))
    sns.heatmap(corr, mask=mask, annot=True, cmap='OrRd', vmin=-1, vmax=1,
                annot_kws={"size": 10}, fmt=".2f")
    plt.title("Feature Correlation Matrix", fontsize=16)
    plt.tight_layout()
    plt.savefig('correlation_matrix.png', dpi=300, bbox_inches='tight')
    plt.show()

enhanced_eda(df)

# 4. Improved Data Preprocessing
def preprocess_data(df):
    """Enhanced preprocessing pipeline"""

    # One-Hot Encoding with drop_first to reduce multicollinearity
    df_encoded = pd.get_dummies(df, columns=['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope'],
                              drop_first=True, dtype='int64')

    # Feature-target split
    X = df_encoded.drop('HeartDisease', axis=1)
    y = df_encoded['HeartDisease']

    # Standardization
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)

    # Train-test split with stratification
    X_train, X_test, y_train, y_test = train_test_split(
        X_scaled, y, test_size=0.2, random_state=42, stratify=y)

    return X_train, X_test, y_train, y_test, X.columns, scaler

X_train, X_test, y_train, y_test, feature_names, scaler = preprocess_data(df)

# 5. Enhanced Model Training with Cross-Validation
def train_and_evaluate_models(X_train, X_test, y_train, y_test, feature_names):
    """Train and evaluate multiple models with cross-validation"""

    models = {
        'Decision Tree': DecisionTreeClassifier(max_depth=5, random_state=42),
        'AdaBoost': AdaBoostClassifier(n_estimators=100, random_state=42),
        'KNN': KNeighborsClassifier(n_neighbors=3),
        'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42)  # New model for comparison
    }

    results = []

    for name, model in models.items():
        # Cross-validation
        cv_scores = cross_val_score(model, X_train, y_train, cv=StratifiedKFold(n_splits=5), scoring='accuracy')

        # Training
        model.fit(X_train, y_train)

        # Evaluation
        y_pred = model.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        roc_auc = roc_auc_score(y_test, model.predict_proba(X_test)[:, 1])

        # Store results
        results.append({
            'Model': name,
            'CV Mean Accuracy': np.mean(cv_scores),
            'CV Std': np.std(cv_scores),
            'Test Accuracy': accuracy,
            'ROC AUC': roc_auc
        })

        # Generate comprehensive classification report
        print(f"\n=== {name} ===")
        print(classification_report(y_test, y_pred))

        # Feature Importance (for tree-based models)
        if hasattr(model, 'feature_importances_'):
            plt.figure(figsize=(10, 6))
            importances = model.feature_importances_
            indices = np.argsort(importances)[::-1]

            plt.title(f"Feature Importance - {name}")
            plt.barh(range(len(indices)), importances[indices], color='#D90404', align='center')
            plt.yticks(range(len(indices)), [feature_names[i] for i in indices])
            plt.xlabel('Relative Importance')
            plt.tight_layout()
            plt.savefig(f'feature_importance_{name.lower().replace(" ", "_")}.png',
                       dpi=300, bbox_inches='tight')
            plt.show()

    # Convert results to DataFrame
    results_df = pd.DataFrame(results)

    # Statistical comparison (paired t-test between best two models)
    best_models = results_df.nlargest(2, 'Test Accuracy')['Model'].values
    model1 = models[best_models[0]]
    model2 = models[best_models[1]]

    # Get cross-validation scores for comparison
    cv1 = cross_val_score(model1, X_train, y_train, cv=StratifiedKFold(n_splits=5), scoring='accuracy')
    cv2 = cross_val_score(model2, X_train, y_train, cv=StratifiedKFold(n_splits=5), scoring='accuracy')

    t_stat, p_val = stats.ttest_rel(cv1, cv2)
    print(f"\nStatistical comparison between {best_models[0]} and {best_models[1]}:")
    print(f"Paired t-test p-value: {p_val:.4f}")

    return results_df

model_results = train_and_evaluate_models(X_train, X_test, y_train, y_test, feature_names)

# 6. Enhanced Hyperparameter Tuning
def hyperparameter_tuning(X_train, y_train):
    """Comprehensive hyperparameter optimization"""

    # AdaBoost tuning
    ab_params = {
        'n_estimators': [50, 100, 150, 200],
        'learning_rate': [0.01, 0.1, 0.5, 1.0]
    }

    ab_grid = GridSearchCV(
        AdaBoostClassifier(random_state=42),
        param_grid=ab_params,
        cv=StratifiedKFold(n_splits=5),
        scoring='accuracy',
        n_jobs=-1,
        verbose=1
    )

    ab_grid.fit(X_train, y_train)

    print("\n=== Best AdaBoost Parameters ===")
    print(ab_grid.best_params_)
    print(f"Best CV Accuracy: {ab_grid.best_score_:.4f}")

    return ab_grid.best_estimator_

best_ab = hyperparameter_tuning(X_train, y_train)

# 7. Final Evaluation with Additional Metrics
def final_evaluation(model, X_test, y_test):
    """Comprehensive final evaluation"""

    # Predictions
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_proba)
    report = classification_report(y_test, y_pred)

    print("\n=== Final Model Evaluation ===")
    print(f"Accuracy: {accuracy:.4f}")
    print(f"ROC AUC: {roc_auc:.4f}")
    print("\nClassification Report:")
    print(report)

    # Visualizations
    fig = plt.figure(figsize=(18, 12))
    gs = fig.add_gridspec(2, 2)

    # Confusion Matrix
    ax1 = fig.add_subplot(gs[0, 0])
    cm = ConfusionMatrix(model, ax=ax1, cmap='OrRd')
    cm.fit(X_train, y_train)  # Fit is meaningless here but required by Yellowbrick
    cm.score(X_test, y_test)
    cm.finalize()

    # ROC Curve
    ax2 = fig.add_subplot(gs[0, 1])
    roc = ROCAUC(model, ax=ax2)
    roc.fit(X_train, y_train)
    roc.score(X_test, y_test)
    roc.finalize()

    # Precision-Recall Curve
    ax3 = fig.add_subplot(gs[1, 0])
    pr = PrecisionRecallCurve(model, ax=ax3)
    pr.fit(X_train, y_train)
    pr.score(X_test, y_test)
    pr.finalize()

    # Class Prediction Error
    ax4 = fig.add_subplot(gs[1, 1])
    cpe = ClassPredictionError(model, ax=ax4)
    cpe.fit(X_train, y_train)
    cpe.score(X_test, y_test)
    cpe.finalize()

    plt.tight_layout()
    plt.savefig('final_model_performance.png', dpi=300, bbox_inches='tight')
    plt.show()

final_evaluation(best_ab, X_test, y_test)

# 8. Deployment-Ready Prediction Function
def predict_heart_disease(model, scaler, feature_names):
    """Enhanced prediction function with input validation"""

    print("\n=== Heart Disease Prediction ===")
    print("Please enter the following clinical features:")

    # Get user input with validation
    try:
        age = float(input("Age (years): "))
        resting_bp = float(input("Resting Blood Pressure (mm Hg): "))
        cholesterol = float(input("Serum Cholesterol (mg/dl): "))
        fasting_bs = int(input("Fasting Blood Sugar > 120 mg/dl (0=No, 1=Yes): "))
        max_hr = float(input("Maximum Heart Rate Achieved: "))
        oldpeak = float(input("Oldpeak (ST depression induced by exercise): "))

        # Categorical features
        print("\nCategorical Features:")
        sex = input("Sex (M/F): ").upper()
        chest_pain = input("Chest Pain Type (ATA/NAP/ASY/TA): ").upper()
        resting_ecg = input("Resting ECG (Normal/ST/LVH): ").title()
        exercise_angina = input("Exercise Angina (Y/N): ").upper()
        st_slope = input("ST Slope (Up/Flat/Down): ").title()

        # Create input DataFrame
        input_data = pd.DataFrame({
            'Age': [age],
            'RestingBP': [resting_bp],
            'Cholesterol': [cholesterol],
            'FastingBS': [fasting_bs],
            'MaxHR': [max_hr],
            'Oldpeak': [oldpeak],
            'Sex_M': [1 if sex == 'M' else 0],
            'ChestPainType_NAP': [1 if chest_pain == 'NAP' else 0],
            'ChestPainType_TA': [1 if chest_pain == 'TA' else 0],
            'ChestPainType_ASY': [1 if chest_pain == 'ASY' else 0],
            'RestingECG_ST': [1 if resting_ecg == 'ST' else 0],
            'RestingECG_LVH': [1 if resting_ecg == 'LVH' else 0],
            'ExerciseAngina_Y': [1 if exercise_angina == 'Y' else 0],
            'ST_Slope_Flat': [1 if st_slope == 'Flat' else 0],
            'ST_Slope_Down': [1 if st_slope == 'Down' else 0]
        })

        # Ensure all features are present (some may be missing due to drop_first)
        for col in feature_names:
            if col not in input_data.columns:
                input_data[col] = 0

        # Reorder columns to match training data
        input_data = input_data[feature_names]

        # Scale the input
        scaled_input = scaler.transform(input_data)

        # Make prediction
        prediction = model.predict(scaled_input)
        probability = model.predict_proba(scaled_input)[0][1]

        # Output result
        print("\n=== Prediction Result ===")
        if prediction[0] == 1:
            print(f"High risk of heart disease (Probability: {probability:.2%})")
            print("Recommended actions: Consult a cardiologist immediately.")
        else:
            print(f"Low risk of heart disease (Probability: {1-probability:.2%})")
            print("Recommended actions: Maintain healthy lifestyle with regular checkups.")

    except Exception as e:
        print(f"Error in input: {str(e)}")
        print("Please ensure all inputs are in the correct format.")

# Example usage (uncomment to test):
# predict_heart_disease(best_ab, scaler, feature_names)

# 9. Save model and scaler for deployment
import joblib

# [Tambahkan di akhir kode]
print("\n=== FINAL SUMMARY ===")

# 1. Tampilkan perbandingan model
print("\nModel Performance Comparison:")
print(model_results.sort_values('Test Accuracy', ascending=False))

# 2. Tampilkan parameter terbaik AdaBoost
print("\nBest AdaBoost Parameters:")
print(best_ab.get_params())

# 3. Contoh prediksi otomatis (tanpa input manual)
print("\nExample Prediction with Sample Data:")
sample_data = {
    'Age': 50, 'RestingBP': 140, 'Cholesterol': 250,
    'FastingBS': 0, 'MaxHR': 160, 'Oldpeak': 1.2,
    'Sex_M': 1, 'ChestPainType_NAP': 0, 'ChestPainType_TA': 0,
    'ChestPainType_ASY': 1, 'RestingECG_ST': 0, 'RestingECG_LVH': 0,
    'ExerciseAngina_Y': 0, 'ST_Slope_Flat': 1, 'ST_Slope_Down': 0
}
sample_df = pd.DataFrame([sample_data], columns=feature_names)
# Fill missing columns with 0
for col in feature_names:
    if col not in sample_data:
        sample_df[col] = 0
# Reorder columns to match feature_names
sample_df = sample_df[feature_names]

scaled_sample = scaler.transform(sample_df)
pred = best_ab.predict(scaled_sample)
prob = best_ab.predict_proba(scaled_sample)[0][1]
print(f"Prediction: {'Disease' if pred[0] == 1 else 'Healthy'}")
print(f"Probability: {prob:.2%}")

# 4. Simpan model (tetap dipertahankan)
joblib.dump(best_ab, 'heart_disease_model.pkl')
joblib.dump(scaler, 'scaler.pkl')
print("\nModel and scaler saved for deployment!")
```
